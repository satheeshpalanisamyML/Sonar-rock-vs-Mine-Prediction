SONAR ROCK VS MINE PREDICTION WITH PYTHON

In this situation, a submarine is navigating through waters during a war between two countries. The enemy country has placed mines under the sea, which explode upon contact with any object. Additionally, there are rocks scattered throughout the ocean. The submarine needs to determine whether it's passing over a mine or a rock. To do this, the submarine sends out sonar signals. These signals bounce off objects in the water and return to the submarine. By analyzing the echoes of these signals, the submarine can distinguish between rocks and potentially dangerous mines. Our goal is to develop a model that can accurately predict whether the submarine is passing over a rock or a mine based on these sonar echoes.

# Import libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics
```

# Data collection and Processing

```python
sonar_data=pd.read_csv('sonar data.csv')
sonar_data
```

# Seperating data and labels

```python

X=sonar_data.drop(columns='R',axis=1)
Y=sonar_data['R']
```

```python
X
```

```python
Y
```

# Training and Test data

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

# Model Training-Logistic Regression

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

# Model Evaluation

```python
y_pred=model.predict(X_test)
```

```python

print("Accuracy:", "%.6f" % metrics.accuracy_score(y_test, y_pred))
```
